Script to parse ACL Anthology bibtex. Results of the parsing saved in pickle format for faster loading.

```python
import bibtexparser
import pandas as pd
```

Takes approximately 6 minutes for 57k papers (papers from 2020 to today).

```python
with open('../data/acl_anthology.bib') as bibtex_file:  ## acl_anthology.bib must be downloaded from the ACL Anthology website and placed in the ../data directory for parsing
    bib_database = bibtexparser.load(bibtex_file)
```

```python
# Example of retrieving a specific entry
bib_database.entries_dict['chen-2025-research']
```

```python
print("Number of Papers:", len(bib_database.entries_dict.keys()))
```

```python
# For first 3 keys in bib_database.entries_dict, print the key and title, abstract
for i, key in enumerate(bib_database.entries_dict.keys()):
    if i >= 3:
        break
    entry = bib_database.entries_dict[key]
    print(f"Key: {key}")
    print(f"Title: {entry.get('title', 'N/A')}")
    print(f"Abstract: {entry.get('abstract', 'N/A')}")
    print()
```

```python
# Load the dictionary into a pandas dataframe
acl_bibliography_df = pd.DataFrame.from_dict(bib_database.entries_dict, orient='index')
```

```python
acl_bibliography_df.head(2)
```

```python
# Scorporate the index to make it a column
acl_bibliography_df.reset_index(inplace=True)
acl_bibliography_df.rename(columns={'index': 'bib_key'}, inplace=True)
```

```python
acl_bibliography_df.to_pickle("../data/acl_bibliography.pkl")
```
